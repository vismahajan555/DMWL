# Data Integration

```python
#To read 2 datasets 
df1 = pd.read_csv('Taylor.csv')
df2 = pd.read_csv('Charlie.csv')
```

```python
df1 #display contents of df1
```

```python
df2  #Display contents of df2
```

```python
df1.isna() #To check NaN values in df1
```

```python
#To delete the row containing NaN values
df1 = df1.drop(4).head()
df1
```

```python
df2.isna()  #To check NaN values in df2
```

```python
#To delete the row containing NaN values
df2 = df2.drop(3).head()
```

```python
df2
```

```python
#To integrate/merge two datasets in df3
df3 = pd.concat([df1, df2])
df3
```
